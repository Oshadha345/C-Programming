#### Traversal

**Types of Traversal**

1. Linear Traversal:

   * Forward: Start at the first element and move towards the last.
   * Backward: Start at the last element and move towards the first.

2. Skipping/Stepping Traversal:

  * Visit every nth element, rather than visiting every element sequentially.

3. Two-Pointer Traversal:

   * Use two indices (pointers) to traverse the array from opposite ends or in a specific pattern.

Traversal Using **range()** in Python




```
range(start, stop, step)
```


*   **start**: The starting number of the sequence (inclusive). Defaults to 0.

*   **stop**: The ending number (exclusive).

* **step**: The difference between each number in the sequence. Defaults to 1.




````python
# simplest traversal

array = [1,2,3,4,5]

for num in array:
  print(num)
````

````python
#Prefix Sum Technique: O(n) (for certain operations)

# Compute the prefix sum array, where each element at index i contains the sum of the elements from index 0 to i

def prefix_sum(arr):
  n = len(arr)
  prefix = [0]*(n)

  for i in range(0,n):
    if i==0:
      prefix[i] = arr[i]
    else:
      prefix[i] = prefix[i-1]+arr[i]

  return prefix

arr = [1, 2, 3, 4, 5, 6, 7, 8]
prefix = prefix_sum(arr)
print(prefix)


#The sum of any subarray from index i to j can be computed in O(1) as:
'''
sum(,) = prefix[]  prefix[-1]

'''

# To find the sum of subarray [2, 3, 4]

i, j = 1, 3  # 0-based indexing

subarray_sum = prefix[j] - prefix[i-1]
print(subarray_sum)
````

# Japura Xtreme

A 3x3 magic square is a grid with 3 rows and 3 columns, containing the numbers. The sum of the numbers in each row, column, and both diagonals is the same, called the magic constant.
In this configuration, the sum of the numbers in each row, column, and diagonal equals 18.

In this question you are presented with an incomplete numberset for a magic square; where the empty spaces are denoted by 0

So your task is to fill the incomplete magic square with a given magic constant!!! It's just simple as that 

**Input Format**

18

9,0,7,0,6,8,5,0,3

**Constraints**

magic number: n ;

0 < n < 50

**Output Format**

9,2,7,4,6,8,5,10,3




````python
def solve_magic_square(magic_constant, square):

    def find_missing(a, b, c):
        return magic_constant - (a + b + c)


    for i in range(9):
        if square[i] == 0:

            # chck row

            row = i // 3
            if square[row*3] + square[row*3 + 1] + square[row*3 + 2] < magic_constant:
                square[i] = find_missing(square[row*3], square[row*3 + 1], square[row*3 + 2])
                continue

            # check column

            col = i % 3
            if square[col] + square[col + 3] + square[col + 6] < magic_constant:
                square[i] = find_missing(square[col], square[col + 3], square[col + 6])
                continue

            # check diagnals

            if i in [0, 4, 8] and square[0] + square[4] + square[8] < magic_constant:
                square[i] = find_missing(square[0], square[4], square[8])
                continue

            if i in [2, 4, 6] and square[2] + square[4] + square[6] < magic_constant:
                square[i] = find_missing(square[2], square[4], square[6])

    return ','.join(map(str, square))

# inp
magic_constant = int(input())
print("")
arry = str(input())

list_arry = list(arry)

square = []
for i in list_arry:
    if i !=",":
        square.append(int(i))




# out
print(solve_magic_square(magic_constant, square))
````

````python
def find_words(matrix, word):
    directions = [
        (-1, 0, "north"), (1, 0, "south"), (0, -1, "west"), (0, 1, "east"),
        (-1, -1, "northwest"), (-1, 1, "northeast"), (1, -1, "southwest"), (1, 1, "southeast")
    ]

    def search_word(word):
        for i in range(10):
            for j in range(10):
                if matrix[i][j] == word[0]:
                    for di, dj, direction in directions:
                        if check_direction(word, i, j, di, dj):
                            return f"({i},{j}) {direction}"
        return None

    def check_direction(word, i, j, di, dj):
        for char in word:
            if i < 0 or i >= 10 or j < 0 or j >= 10 or matrix[i][j] != char:
                return False
            i, j = i + di, j + dj
        return True

    results = []
    for word in words:
        result = search_word(word)
        if result:
            results.append(result)

    return results

# Main execution
n = int(input())  # Number of words to find

words = []
for _ in  range(n):
    word = input().split
    words.append(word)

# Read the matrix
matrix = []
for _ in range(10):
    row = input().strip()
    matrix.append(row)

# Find words and print results
results = find_words(matrix, words)
for result in results:
    print(result)
````

````python
#Leetcode Roman to Integer

def romanToInt(s):
        symbol = ['M','D','C','L','X','V','I','IV','IX','XL','XC','CD','CM']
        values = [1000,500,100,50,10,5,1,4,9,40,90,400,900]

        int_value = 0
        for i in range(len(s)):
            if i!=len(s)-1:
                if s[i:i+2] in symbol:

                    int_value += values[symbol.index(s[i:i+2])]
                    if i+1==len(s)-1:
                      int_value += values[symbol.index(s[i+1])]



                    int_value -= values[symbol.index(s[i+1])]


                else:
                    int_value += values[symbol.index(s[i])]


            else:
                if s[i-1:i+1] not in symbol:
                    int_value += values[symbol.index(s[i])]

        return int_value


s = "CMCDXLIV"
print(romanToInt(s))
````

# Divide and Conquer

Steps in divide and conquer:

1. **Divide** : Break the problem into smaller subproblems of the same type.

2. **Conquer** : Solve the subproblems recursively. If they are small enough, solve them directly.

3. **Combine** : Combine the solutions of the subproblems to get the solution to the original problem.

### Binary Search


$ Time Complexity $ : $T(n) = T(n/2) + O(1)$

````python
# recursive method of binary search
def binary_search(arr, left, right, target):

  if left > right:
    return False

  mid = left + (right-left)//2

  if arr[mid] == target:
    return mid
  elif arr[mid]<target:
    return binary_search(arr,mid+1,right,target)
  else:
    return binary_search(arr,left,mid-1,target)


arr = list(map(int,input().split()))
target = int(input())

print(binary_search(arr,0,len(arr)-1,target))
````

````python
#binary search on rotated sorted list


def bin_rot_search(arr,target):
  l, r = 0, len(arr)-1

  while l <= r:
    mid = l + (r-l)//2

    if arr[mid] == target:
      return mid

    else:
      if arr[mid] > target:
        if arr[l]<=target or arr[mid]<arr[l]:
          r = mid - 1
        else:
          l = mid + 1
      else:
        if arr[r]>=target or arr[mid]>arr[r]:
          l = mid + 1
        else:
          r = mid - 1

  return False

arr = list(map(int,input().split()))
target = int(input())

print(bin_rot_search(arr,target))
````

````python
#iterative method of binary search

def bin_search(arr,target):
  l, r = 0, len(arr)-1

  while l <= r:
    mid = l + (r-l)//2

    if target > arr[mid]:
      l = mid + 1
    elif target < arr[mid]:
      r = mid - 1
    else:
      return mid
  return False

arr = list(map(int,input().split()))
target = int(input())

print(bin_search(arr,target))
````

````python
# Bubble sort----> time complexity o(n**2)

array = [2,1,7,3,4,5,6]

n = len(array)

for i in range(n):
  swapped = False
  for j in range(len(array)-i-1):
    if array[j]>array[j+1]:
      array[j], array[j+1] = array[j+1], array[j]
      swapped = True
  if not swapped:
    break

print(array)
````

### Selection Sort

````python
# Selection sort---> time complexity O(n**2)

def selectionSort(arr):

  for i in range(len(arr)-1):
    min_index = i
    for j in range(i+1,len(arr)):
      if arr[j]<arr[min_index]:
        min_index = j
    min_ = arr.pop(min_index)
    arr.insert(i,min_)

  return arr


arr = [7,12,9,11,3]
print(selectionSort(arr))
````

````python
# Selection sort swap

def selection_sort(arr):
  n = len(arr)

  for i in range(n-1):
    min_index = i
    for j in range(i+1,n):
      if arr[j]<arr[min_index]:
        min_index = j
    arr[i], arr[min_index] = arr[min_index], arr[i]

  return arr

arr = [7,12,9,11,3]
print(selection_sort(arr))
````

### Insertion Sort

````python
# Insertion sort

def insertionSort(arr):
  n = len(arr)

  for i in range(n-1):
    for j in range(i+1,0,-1):
      if arr[j]<arr[j-1]:
        arr[j], arr[j-1] = arr[j-1], arr[j]
      else:
        break

  return arr

arr = [7,12,9,11,3]
print(insertionSort(arr))
````

````python
# optimized insertion sort

def insertion_sort(arr):
  n = len(arr)
  result = []

  for i in arr:
    result.append(i)

    for j in range(len(result)-1,0,-1):
      if result[j-1]>result[j]:
        result[j-1], result[j] = result[j], result[j-1]
      else:
        break

  return result

arr = [7,12,9,11,3]
print(insertion_sort(arr))
````

````python
#insertion sort improved

def opt_insertion_sort(arr):
  n = len(arr)

  for i in range(1,n):
    insert_index = i
    current_value = arr[i]

    for j in range(i-1,-1,-1):
      if arr[j]>current_value:
        arr[j+1] = arr[j]
        insert_index = j
      else:
        break
    arr[insert_index] = current_value

  return arr

arr = [7,12,9,11,3]
print(opt_insertion_sort(arr))
````

### Quick Sort

````python
#Quick sort---> time complexity best-case->O(nlogn)~for random arrays, worst-case->O(n**2)~for sorted arrays

def quickSort(arr,low=0,high=None):
  if high is None:
    high = len(arr)-1

  if low < high:
    pivot = partition(arr,low,high)
    quickSort(arr,low,pivot-1)
    quickSort(arr,pivot+1,high)

  return arr


def partition(arr,low,high):
  pivot = arr[high]
  i = low - 1
  for j in range(low,high):
    if arr[j]<=pivot:
      i+=1
      arr[i], arr[j] = arr[j], arr[i]
  arr[i+1], arr[high] = arr[high], arr[i+1]

  return i+1



arr = [7,12,9,11,3]
print(quickSort(arr))
````

````python
def Quick_sort(arr):
  if len(arr)<=1:
    return arr
  else:
    pivot = arr[0]
    less = [x for x in arr[1:] if x<=pivot]
    greater = [x for x in arr[1:] if x>pivot]

    return Quick_sort(less) + [pivot] + Quick_sort(greater)

arr = [7,12,9,11,3]
print(Quick_sort(arr))
````

````python
#time complexity --> O(nlogn)

def merge(left,right):
  l = len(left)
  r = len(right)
  i = j = 0

  result = []

  while i<l and j<r:
    if left[i]<=right[j]:
      result.append(left[i])
      i+=1
    else:
      result.append(right[j])
      j+=1

  while i<l:
    result.append(left[i])
    i+=1


  while j<r:
    result.append(right[j])
    j+=1


  return result


def mergeSort(arr):
  if len(arr)>1:

    mid = len(arr)//2

    left = arr[:mid]
    right = arr[mid:]

    left = mergeSort(left)
    right = mergeSort(right)

    return merge(left,right)

  return arr

arr = [7,12,9,11,3]
print(mergeSort(arr))
````

### Tim Sort(Hybrid Sort)


````python
def InsertionSort(arr,left,right):
  for i in range(left+1,right+1):
    current_value = arr[i]
    insert_index = i

    for j in range(i-1,left-1,-1):
      if current_value<arr[j]:
        arr[j+1] = arr[j]
        insert_index = j
      else:
        break
      arr[insert_index] = current_value

def Merge(arr,l,m,r):
  left = arr[l:m+1]
  right = arr[m+1:r+1]
  i, j = 0, 0
  k = l

  while i<len(left) and j<len(right):
    if left[i]<=right[j]:
      arr[k] = left[i]
      i+=1
    else:
      arr[k] = right[j]
      j+=1
    k+=1

  while i<len(left):
    arr[k] = left[i]
    i+=1
    k+=1

  while j<len(right):
    arr[k] = right[j]
    j+=1
    k+=1


def timSort(arr):
  n = len(arr)
  minrun = 32

  #perform insertion sort on runs

  for start in range(0,n,minrun):
    end = min(n-1,start+minrun-1)
    InsertionSort(arr,start,end)

  #merge runs
  size = minrun

  while size < n:

    for left in range(0,n,size*2):
      mid = min(n-1,left+size-1)
      right = min(n-1,left+size*2-1)
      if mid < right :
        Merge(arr,left,mid,right)
    size*=2


arr = [7,12,9,11,3]
timSort(arr)
print(arr)
````

### Counting Sort

````python
#Count sort ---> time complexity O(n+k)

def countingSort(arr):
  n = len(arr)
  max_ = max(arr)
  min_ = min(arr)

  count = [0]*(max_-min_+1)

  for i in range(n):
    count[arr[i]-min_]+=1

  k=0

  for j in range(len(count)):
    while count[j]!=0:
      arr[k] = j+min_
      k+=1
      count[j]-=1

  return arr

arr = [7,12,9,11,3]
print(countingSort(arr))
````

````python
#counting sort cumulative approach

def Counting_sort(arr):
  n = len(arr)
  maximum  = max(arr)
  minimum = min(arr)

  count = [0]*(maximum-minimum+1)
  result = [0]*n

  for i in range(n):
    count[arr[i]-minimum]+=1

  for j in range(1,len(count)):
    count[j] += count[j-1]


  for k in range(n-1,-1,-1):
    result[count[arr[k]-minimum]-1] = arr[k]
    count[arr[k]-minimum] -= 1

  return result

arr = [7,12,9,11,3]
print(Counting_sort(arr))
````

````python
def counting_sort(arr, max_val):
    count = [0] * (max_val + 1)  # Create count array for the range of elements
    output = [0] * len(arr)  # Output array for sorted elements

    # Count the occurrence of each element
    for num in arr:
        count[num] += 1

    # Update the count array to store cumulative counts
    for i in range(1, len(count)):
        count[i] += count[i - 1]

    # Build the output array by placing elements in their sorted positions
    for num in reversed(arr):
        output[count[num] - 1] = num
        count[num] -= 1

    return output

# Example usage:
arr = [3, 6, 4, 1, 3, 4, 1, 4]
max_val = max(arr)
sorted_arr = counting_sort(arr, max_val)
print(sorted_arr)
````

###Radix Sort

````python
# radix sort ---> time complexity O()

def counting_sort(arr, digit):
    count = [0] * (10)
    output = [0] * len(arr)


    for num in arr:
      index = (num//(10**(digit-1))) % 10
      count[index] += 1

    for i in range(1, len(count)):
        count[i] += count[i - 1]

    for num in reversed(arr):
        index = (num//(10**(digit-1))) % 10
        output[count[index] - 1] = num
        count[index] -= 1

    return output


def radixSort(arr):
  n = len(arr)
  max_ = max(arr)

  length = len(str(max_))

  for i in range(1,length+1):
    arr = counting_sort(arr,i)

  return arr

arr = [170, 45, 75, 90, 802, 24, 2, 66]
print(radixSort(arr))
````
